# VISUALIZATION 1- GROUPED BAR CHART

```python
# Create a grouped bar chart for male vs female population
plt.figure(figsize=(15, 8))

# Select top 10 states by population for clarity
top_states = df.sort_values('Population[50]', ascending=False).head(10).copy()
states = top_states['State/UT']

# Set bar positions
x = np.arange(len(states))
width = 0.35

# Create bars
bars1 = plt.bar(x - width/2, top_states['Male']/1e6, width, 
                label='Male', color='#1f77b4', alpha=0.8, edgecolor='black')
bars2 = plt.bar(x + width/2, top_states['Female']/1e6, width, 
                label='Female', color='#ff7f0e', alpha=0.8, edgecolor='black')

# Customize
plt.xlabel('State/UT', fontsize=12, fontweight='bold')
plt.ylabel('Population (in Millions)', fontsize=12, fontweight='bold')
plt.title('Male vs Female Population in Top 10 Indian States', 
          fontsize=16, fontweight='bold', pad=20)
plt.xticks(x, states, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                 f'{height:.1f}M', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('gender_distribution_states.png', dpi=300, bbox_inches='tight')
plt.show()
```

### DATA CLEANING AND COLUMN RENAMING

```python
# Cell 1: Clean and rename columns
print("=== CLEANING COLUMN NAMES ===")

# Create a copy to work with
df_clean = df.copy()

# Let's see all column names first
print("Original column names:")
for col in df_clean.columns:
    print(f"  '{col}'")

# Rename columns properly
column_renames = {
    'Population[50]': 'Population',
    'Percent (%)': 'Percent',
    'Difference between male and female': 'Gender_Difference',
    'Sex ratio': 'Sex_Ratio',
    'Rural[51]': 'Rural',
    'Urban[51]': 'Urban',
    'Area[52] (km2)': 'Area_km2',
    'Density (per km2)': 'Density'
}

# Apply renaming
df_clean.rename(columns=column_renames, inplace=True)

print("\nCleaned column names:")
for col in df_clean.columns:
    print(f"  '{col}'")

print("\nFirst 2 rows after cleaning:")
display(df_clean.head(2))
```

# VISUALISATION 2- HISTOGRAM.

```python

print("=== CREATING GENDER DISTRIBUTION BAR CHART ===")

plt.figure(figsize=(15, 8))

# Select top 10 states by population for clarity
top_states = df_clean.sort_values('Population', ascending=False).head(10).copy()
states = top_states['State/UT']

print(f"Top 10 states by population:")
for i, (state, pop) in enumerate(zip(states, top_states['Population']/1e6), 1):
    print(f"{i:2}. {state}: {pop:.1f} million")

# Set bar positions
x = np.arange(len(states))
width = 0.35

# Create bars (convert to millions)
bars1 = plt.bar(x - width/2, top_states['Male']/1e6, width, 
                label='Male', color='#1f77b4', alpha=0.8, edgecolor='black')
bars2 = plt.bar(x + width/2, top_states['Female']/1e6, width, 
                label='Female', color='#ff7f0e', alpha=0.8, edgecolor='black')

# Customize
plt.xlabel('State/UT', fontsize=12, fontweight='bold')
plt.ylabel('Population (in Millions)', fontsize=12, fontweight='bold')
plt.title('Male vs Female Population in Top 10 Indian States', 
          fontsize=16, fontweight='bold', pad=20)
plt.xticks(x, states, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                 f'{height:.1f}M', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('gender_distribution_bar_chart.png', dpi=300, bbox_inches='tight')
plt.show()

print("âœ… Bar chart created successfully!")
```

```python

import matplotlib.pyplot as plt
import numpy as np

print("=== CREATING HISTOGRAMS ===")

# Check if we have the right columns
print("Available columns for histogram:")
print([col for col in df_clean.columns if 'Population' in col or 'Sex' in col])

plt.figure(figsize=(14, 6))

# Create two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Subplot 1: Total Population Distribution
# Convert population to millions for better readability
population_millions = df_clean['Population'] / 1e6

ax1.hist(population_millions, bins=15, color='skyblue', 
         edgecolor='black', alpha=0.7)
ax1.axvline(population_millions.mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Mean: {population_millions.mean():.1f}M')
ax1.set_xlabel('Population (Millions)', fontsize=11)
ax1.set_ylabel('Number of States/UTs', fontsize=11)
ax1.set_title('Distribution of State Populations', fontsize=13, fontweight='bold')
ax1.legend()
ax1.grid(alpha=0.3)

# Subplot 2: Sex Ratio Distribution
ax2.hist(df_clean['Sex_Ratio'], bins=15, color='lightcoral', 
         edgecolor='black', alpha=0.7)
ax2.axvline(df_clean['Sex_Ratio'].mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Mean: {df_clean["Sex_Ratio"].mean():.1f}')
ax2.axvline(1000, color='green', linestyle='-', 
            linewidth=1.5, alpha=0.5, label='Equal Ratio (1000)')
ax2.set_xlabel('Sex Ratio (Females per 1000 Males)', fontsize=11)
ax2.set_ylabel('Number of States/UTs', fontsize=11)
ax2.set_title('Distribution of Sex Ratios', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(alpha=0.3)

plt.suptitle('Population Analysis of Indian States/UTs', 
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('population_histograms.png', dpi=300, bbox_inches='tight')
plt.show()

print("âœ… Histograms created successfully!")
```

#### SIMPLE ANALYSIS SUMMARY

```python
# Cell 4: Basic Statistics Summary
print("=== BASIC STATISTICS SUMMARY ===")
print(f"Total number of states/UTs: {len(df_clean)}")
print(f"Total population of India (from data): {df_clean['Population'].sum():,}")
print(f"Total male population: {df_clean['Male'].sum():,}")
print(f"Total female population: {df_clean['Female'].sum():,}")
print(f"Overall sex ratio: {(df_clean['Female'].sum()/df_clean['Male'].sum()*1000):.1f}")
print(f"Average sex ratio across states: {df_clean['Sex_Ratio'].mean():.1f}")

print("\n=== STATES WITH HIGHEST POPULATION ===")
top_5 = df_clean.nlargest(5, 'Population')[['State/UT', 'Population', 'Sex_Ratio']]
top_5['Population_Millions'] = top_5['Population'] / 1e6
print(top_5[['State/UT', 'Population_Millions', 'Sex_Ratio']].to_string(index=False))

print("\n=== GENDER GAP ANALYSIS ===")
df_clean['Gender_Gap_Millions'] = (df_clean['Male'] - df_clean['Female']) / 1e6
print(f"Average gender gap (male-female): {df_clean['Gender_Gap_Millions'].mean():.2f} million")
print(f"State with highest male surplus: {df_clean.loc[df_clean['Gender_Gap_Millions'].idxmax(), 'State/UT']}")
print(f"State with highest female surplus: {df_clean.loc[df_clean['Gender_Gap_Millions'].idxmin(), 'State/UT']}")
```

### VISUALIZATION OF GENDER RATIO DISTRIBUTION 

```python

print("=== ADDITIONAL VISUALIZATION: GENDER RATIO DISTRIBUTION ===")

plt.figure(figsize=(12, 6))

# Create a more detailed histogram for Sex Ratio
# Sex Ratio is a CONTINUOUS variable (can range from 800 to 1200, for example)

# Calculate statistics
mean_ratio = df_clean['Sex_Ratio'].mean()
median_ratio = df_clean['Sex_Ratio'].median()
std_ratio = df_clean['Sex_Ratio'].std()

# Create histogram
n, bins, patches = plt.hist(df_clean['Sex_Ratio'], bins=20, 
                           color='lightgreen', edgecolor='black', 
                           alpha=0.7, density=False)

# Add reference lines
plt.axvline(mean_ratio, color='red', linestyle='--', linewidth=2, 
            label=f'Mean: {mean_ratio:.1f}')
plt.axvline(median_ratio, color='blue', linestyle='--', linewidth=2, 
            label=f'Median: {median_ratio:.1f}')
plt.axvline(1000, color='green', linestyle='-', linewidth=1.5, alpha=0.5, 
            label='Equal (1000)')

# Customize
plt.xlabel('Sex Ratio (Females per 1000 Males)', fontsize=12)
plt.ylabel('Number of States/UTs', fontsize=12)
plt.title('Distribution of Sex Ratios Across Indian States/UTs\n(A Continuous Variable)', 
          fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)

# Add statistics text box
stats_text = f"""Statistics:
Total States: {len(df_clean)}
Mean Ratio: {mean_ratio:.1f}
Median Ratio: {median_ratio:.1f}
Std Deviation: {std_ratio:.1f}
Min Ratio: {df_clean['Sex_Ratio'].min():.1f}
Max Ratio: {df_clean['Sex_Ratio'].max():.1f}"""
plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
         fontsize=10)

plt.tight_layout()
plt.savefig('sex_ratio_distribution_histogram.png', dpi=300, bbox_inches='tight')
plt.show()

print("âœ… Additional histogram created!")
print("   This shows distribution of a CONTINUOUS variable (Sex Ratio)")
```
